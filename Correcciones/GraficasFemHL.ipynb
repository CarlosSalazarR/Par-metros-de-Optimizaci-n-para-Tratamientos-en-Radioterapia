```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
from matplotlib import pyplot as plt
```

```python
A1 = pd.read_csv("20_Med.txt") #Plantilla A
A2 = pd.read_csv("20Pareto.txt") #B
A3 = pd.read_csv("20ParetoB.txt") #C

DICCS1 = [A1]
DICCS2 = [A2]
DICCS3 = [A3]

L1 = 21
L2 = 31
for i in range(L1, L2+1):
    if i!=25:
        B1 =  pd.read_csv(str(i)+"_Med.txt")
        A1 = A1 + B1
        B2 =  pd.read_csv(str(i)+"Pareto.txt")
        A2 = A2 + B2        #hacer lista de diccionarios
        B3 =  pd.read_csv(str(i)+"ParetoB.txt")
        A3 = A3 + B3        #hacer lista de diccionarios
        DICCS1.append(B1)
        DICCS2.append(B2)
        DICCS3.append(B3)
A1 = A1.dropna()
A2 = A2.dropna()
A3 = A3.dropna()

promedio1 = A1/(L2-L1 + 1)
promedio2 = A2/(L2-L1 + 1)
promedio3 = A3/(L2-L1 + 1)
```

```python
desviacion1 = []

for j in range(len(promedio1["Dose_[cGy]"])):
    ratios1 = []
    for i in DICCS1:
        ratios1.append(i["Ratio_of_Total_Structure_Volume_[%]"][j]) #j esimo valor
    desviacion1.append(np.std(ratios1))
promedio1["Desviacion_Estandar"] = desviacion1 #tomar el promedio de los promedios?
```

```python
desviacion2 = []

for j in range(len(promedio2["Dose_[cGy]"])):
    ratios2 = []
    for i in DICCS2:
        ratios2.append(i["Ratio_of_Total_Structure_Volume_[%]"][j]) #j esimo valor
    desviacion2.append(np.std(ratios2))
promedio2["Desviacion_Estandar"] = desviacion2 #tomar el promedio de los promedios?
```

```python
desviacion3 = []

for j in range(len(promedio3["Dose_[cGy]"])):
    ratios3 = []
    for i in DICCS3:
        ratios3.append(i["Ratio_of_Total_Structure_Volume_[%]"][j]) #j esimo valor
    desviacion3.append(np.std(ratios3))
promedio3["Desviacion_Estandar"] = desviacion3 #tomar el promedio de los promedios?
```

```python
from matplotlib.pyplot import figure
```

```python
vol_prom1 = promedio1["Ratio_of_Total_Structure_Volume_[%]"]
vol_prom2 = promedio2["Ratio_of_Total_Structure_Volume_[%]"]
vol_prom3 = promedio3["Ratio_of_Total_Structure_Volume_[%]"]

dosis1 = promedio1["Relative_dose_[%]"]
dosis2 = promedio2["Relative_dose_[%]"]
dosis3 = promedio3["Relative_dose_[%]"]

desvmed1= desviacion1/np.sqrt(L2+1-L1) #Desviacion estandar de la media
desvmed2= desviacion2/np.sqrt(L2+1-L1) #Desviacion estandar de la media
desvmed3= desviacion3/np.sqrt(L2+1-L1) #Desviacion estandar de la media

fig, ax = plt.subplots(1)
figure(figsize=(20, 20), dpi=80)

ax.plot(dosis1, vol_prom1, lw=2, label='Femoral Head L, plantilla A', color='blue')
ax.plot(dosis2, vol_prom2, lw=2, label='Femoral Head L, plantilla B', color='red')
ax.plot(dosis3, vol_prom3, lw=2, label='Femoral Head L, plantilla C', color='green')

#ax.plot(t, mu2, lw=2, label='mean population 2', color='yellow')
#ax.fill_between(dosis, vol_prom+desviacion, vol_prom-desviacion, facecolor='blue', alpha=0.5)
ax.fill_between(dosis1, vol_prom1+desvmed1, vol_prom1-desvmed1, facecolor='blue', alpha=0.5)
ax.fill_between(dosis2, vol_prom2+desvmed2, vol_prom2-desvmed2, facecolor='red', alpha=0.5)
ax.fill_between(dosis3, vol_prom3+desvmed3, vol_prom3-desvmed3, facecolor='green', alpha=0.5)

#ax.fill_between(t, mu2+sigma2, mu2-sigma2, facecolor='yellow', alpha=0.5)
ax.set_title(r'')
ax.legend(loc='upper left')
ax.set_xlabel('Relative Dosis [%]')
ax.set_ylabel('Relative Volume [%]')
ax.grid()
plt.show()
```
